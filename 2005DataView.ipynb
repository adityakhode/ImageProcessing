```python
#Election data analysys

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")
df
```

```python
# Ensure AGE is integer
df['AGE'] = df['AGE'].astype(int)

# Total count per age
age_count = df['AGE'].value_counts().sort_index()

# Gender-wise count per age
male_count = df[df['SEX'] == 'M']['AGE'].value_counts().sort_index()
female_count = df[df['SEX'] == 'F']['AGE'].value_counts().sort_index()

# Ensure all ages exist (21–95)
ages = list(range(21, 96))

total = [age_count.get(a, 0) for a in ages]
male = [male_count.get(a, 0) for a in ages]
female = [female_count.get(a, 0) for a in ages]
```

```python
plt.figure(figsize=(15, 5))

# Bar plot for total population
plt.bar(ages, total, color="#4facfe", alpha=0.6, label="Total")

# Line plot for males
plt.plot(ages, male, color="red", linewidth=2, marker='o', label="Male")

# Line plot for females
plt.plot(ages, female, color="green", linewidth=2, marker='o', label="Female")

plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age-wise Population with Gender Distribution")

# Reduce x-axis clutter
plt.xticks(range(21, 96, 1))

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
